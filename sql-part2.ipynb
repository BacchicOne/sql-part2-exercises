To get started with the exercises, go ahead and take a look at RideShareDB. This database was created with data from a dataset on [Kaggle](https://www.kaggle.com/fivethirtyeight/uber-pickups-in-new-york-city). The dataset was made using data obtained from the NYC Taxi and Limosine Commission or TLC. TLC tracks this data by base and even rideshare companies such as Uber and Lyft have assigned bases that each pickup must be attributed to. ## String Functions

Write a query that returns the last 3 characters of the base name in `other-FHV-services-jan-july15`. Do you see anything that is a common abbreviation at the end of business names?

```sql
SELECT RIGHT(Base_Name, 3) AS Name 
FROM dbo.other_FHV_services_jan_aug_2015
ORDER BY Base_Name
```

## Date Functions

For ``other-FHV-services-jan-july15``, display the name of month of the pickup date and order in descending order. Is the result of this query what you expected?

```sql
SELECT DATENAME(month, Pick_Up_Date)
FROM dbo.other_FHV_services_jan_aug_2015
ORDER BY Pick_Up_Date DESC;
```

Using `other-FHV-services-jan-july15`, write a query that returns the day number of the pickup date and month name and orders it in ascending order by day number.

```sql
SELECT 
    DATENAME(day, Pick_Up_Date) AS PickUpDay,
    DATENAME(month, Pick_Up_Date) AS PickUpMonth
FROM dbo.other_FHV_services_jan_aug_2015
ORDER BY PickUpDay ASC
```
